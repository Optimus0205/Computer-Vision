```python
import numpy as np
import cv2
```

```python
image_path='/content/Vertical Half Image.jpg'

image=cv2.imread(image_path)
image
```

```python
image_path='/content/Vertical Half Image.jpg'

image=cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
image
```

Creating Sobel Vertical

```python
sobel_vertical=np.array([[-1, 0, 1],
                         [-2, 0, 2],
                         [-1, 0, 1]])

sobel_vertical
```

```python
image.shape
```

```python
rows,cols= image.shape
rows,cols
```

```python
k=sobel_vertical.shape[0]
k
```

```python
output_matrix = np.zeros((rows-k+1, cols-k+1))
output_matrix
```

```python
for i in range(1, rows-1):
    for j in range(1, cols-1):
        region=image[i-1:i+2, j-1:j+2]
        output_matrix[i-1, j-1]=np.sum(region*sobel_vertical)
        print(output_matrix)
```

```python
output_matrix
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))

# original image

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')


plt.subplot(1,2,2)
plt.title('Sobel Conv Edge Output')
plt.imshow(output_matrix, cmap='gray')
plt.axis('off')

plt.show()
```
