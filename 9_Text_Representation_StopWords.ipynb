```python
import spacy

from spacy.lang.en.stop_words import STOP_WORDS

STOP_WORDS
```

```python
len(STOP_WORDS)
```

```python
nlp=spacy.load('en_core_web_sm')

doc=nlp('We just opened our wings, the flying part is coming soon.')

for token in doc:
    if token.is_stop:
      print(token)
```

```python
def preprocess(text):
    doc=nlp(text)

    no_stop_words=[token.text for token in doc if not token.is_stop]
    return no_stop_words
```

```python
preprocess('We just opened our wings, the flying part is coming soon.')
```

```python
def preprocess1(text):
    doc=nlp(text)

    no_stop_words=[token.text for token in doc if not token.is_stop and not token.is_punct]
    return no_stop_words
```

```python
preprocess1('We just opened our wings, the flying part is coming soon.')
```

```python
preprocess1('The other is not other but your divine brother.')
```

```python
preprocess1('Musk wants time to prepare for a trial over his')
```

```python
def preprocess2(text):
    doc=nlp(text)

    no_stop_words=[token.text for token in doc if not token.is_stop and not token.is_punct]
    return ''.join(no_stop_words)
```

```python
df['contents_new']=df['contents'].apply(preprocess2)

df.head()
```
